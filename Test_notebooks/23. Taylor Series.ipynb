### Part 1: Taylor Series Expansions

1. **For $ f(x) = \cos(x) $, centered at $ x = 0 $:**
   The Taylor series expansion is:
   $
   f(x) = \cos(x) = \sum_{n=0}^\infty \frac{(-1)^n x^{2n}}{(2n)!}.
   $
   Expanding up to the 4th degree:
   $
   f(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} + \mathcal{O}(x^6).
   $
   Simplifying:
   $
   f(x) = 1 - \frac{x^2}{2} + \frac{x^4}{24}.
   $

2. **For $ h(x) = \frac{1}{1-x} $, centered at $ x = 0 $:**
   The Taylor series expansion is:
   $
   h(x) = \sum_{n=0}^\infty x^n.
   $
   Expanding up to the 4th degree:
   $
   h(x) = 1 + x + x^2 + x^3 + x^4.
   $

3. **For $ g(x) = \sin(x) $, centered at $ x = \pi $:**
   Perform a Taylor expansion around \( x = \pi \). Let \( u = x - \pi \), so $ g(x) = \sin(\pi + u) = -\sin(u) $. The Taylor series for $ \sin(u) $ is:
   $
   \sin(u) = u - \frac{u^3}{3!} + \mathcal{O}(u^5).
   $
   Substituting back $ u = x - \pi $:
   $
   g(x) = -(x - \pi) + \frac{(x - \pi)^3}{6}.
   $
   So the expansion is:
   $
   g(x) = - (x - \pi) + \frac{(x - \pi)^3}{6}.
   $

### TASK 2: Find the Tangent Line for $ f(x) = e^{\sin(x)} $ at $ x_0 = \pi $

#### Step 1: Tangent Line Formula
The general formula for a tangent line is:
$
y = f'(x_0)(x - x_0) + f(x_0)
$
Here:
- $ f'(x_0) $ is the derivative of $ f(x) $ evaluated at $ x_0 $,
- $ f(x_0) $ is the function value at $ x_0 $.

---

#### Step 2: Evaluate $ f(x_0) $ and $ f'(x_0) $

1. **Compute $ f(x_0) $:**
   Given $ f(x) = e^{\sin(x)} $,
   $
   f(x_0) = f(\pi) = e^{\sin(\pi)} = e^0 = 1
   $

2. **Compute $ f'(x_0) $:**
   The derivative of $ f(x) = e^{\sin(x)} $ is:
   $
   f'(x) = e^{\sin(x)} \cdot \cos(x)
   $
   At \( x_0 = \pi \):
   $
   f'(x_0) = e^{\sin(\pi)} \cdot \cos(\pi) = e^0 \cdot (-1) = -1
   $

---

#### Step 3: Tangent Line Equation

Using the tangent line formula:
$
y = f'(x_0)(x - x_0) + f(x_0)
$
Substitute $ f'(x_0) = -1 $, $ x_0 = \pi $, and $ f(x_0) = 1 $:
$
y = -1 \cdot (x - \pi) + 1
$
Simplify:
$
y = -x + \pi + 1
$

---

#### Final Tangent Line:
The equation of the tangent line is:
$
y = -x + \pi + 1
$
