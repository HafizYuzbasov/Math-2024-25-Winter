### TASK 2

To calculate the integrals of $ f(x) = 2x + 1 $ and $ g(x) = x^2 $ over the interval $[0, \pi]$, we proceed as follows:

---

### Integral of $ f(x) = 2x + 1 $:
$
\int_{0}^{\pi} (2x + 1) \, dx
$

1. Compute the indefinite integral:
$
\int (2x + 1) \, dx = x^2 + x + C
$

2. Evaluate the definite integral:
$
\int_{0}^{\pi} (2x + 1) \, dx = \left[ x^2 + x \right]_0^\pi
$

3. Substitute the limits:
$
= \left( \pi^2 + \pi \right) - \left( 0^2 + 0 \right)
$

4. Final result:
$
\int_{0}^{\pi} (2x + 1) \, dx = \pi^2 + \pi
$

### Integral of $ g(x) = x^2 $:
$
\int_{0}^{\pi} x^2 \, dx
$

1. Compute the indefinite integral:
$
\int x^2 \, dx = \frac{x^3}{3} + C
$

2. Evaluate the definite integral:
$
\int_{0}^{\pi} x^2 \, dx = \left[ \frac{x^3}{3} \right]_0^\pi
$

3. Substitute the limits:
$
= \frac{\pi^3}{3} - \frac{0^3}{3}
$

4. Final result:
$
\int_{0}^{\pi} x^2 \, dx = \frac{\pi^3}{3}
$

### Step 1: Understand the Problem
The region lies between $x = 1$ and $x = 2$, and the curve $y = x^2 + 1$ is above the line $y = 0$. Thus, the area is determined by integrating $y = x^2 + 1$ with respect to $x$ over the interval $[1, 2]$.

---

### Step 2: Set Up the Integral
The area $A$ is given by:
$
A = \int_{1}^{2} \left( x^2 + 1 \right) dx
$

---

### Step 3: Compute the Integral
Break it into two parts:
$
\int_{1}^{2} \left( x^2 + 1 \right) dx = \int_{1}^{2} x^2 \, dx + \int_{1}^{2} 1 \, dx
$

1. Compute $\int_{1}^{2} x^2 dx$:
$
\int x^2 dx = \frac{x^3}{3} \quad \text{so evaluate: } \left[ \frac{x^3}{3} \right]_1^2 = \frac{2^3}{3} - \frac{1^3}{3} = \frac{8}{3} - \frac{1}{3} = \frac{7}{3}.
$

2. Compute $\int_{1}^{2} 1 dx$:
$
\int 1 dx = x \quad \text{so evaluate: } \left[ x \right]_1^2 = 2 - 1 = 1.
$

---

### Step 4: Combine the Results
Add the results:
$
A = \frac{7}{3} + 1 = \frac{7}{3} + \frac{3}{3} = \frac{10}{3}.
$

---

### Final Answer:
The area of the region is:
$
\boxed{\frac{10}{3}}

### TASK 4. **Calculate the area under the sine curve over the interval $[0, \pi]$:**

We need to compute:

$
P = \int_0^\pi \sin(x) \, dx
$

**Solution:**

The integral of $\sin(x)$ is $-\cos(x)$. Evaluating:

$
P = \int_0^\pi \sin(x) \, dx = \big[-\cos(x)\big]_0^\pi = -\cos(\pi) + \cos(0)
$

$
P = -(-1) + 1 = 2
$

### Final Answers:
 **Area under sine curve:** \(P = 2\)

### TASK 6. **Find the distance of the moving particle between $t = 0$ and $t = 2$:**

The position function is:

$
x(t) = 3t^2 - 6t + 1
$

The distance traveled is the integral of the absolute value of the velocity. First, compute the velocity:

$
v(t) = x'(t) = 6t - 6
$

Find when $v(t) = 0$:

$
6t - 6 = 0 \quad \Rightarrow \quad t = 1
$

Now split the interval $[0, 2]$ into two parts: $[0, 1]$ and $[1, 2]$. On $[0, 1]$, $v(t) \leq 0$, and on $[1, 2]$, $v(t) \geq 0$.

$
\text{Distance} = \int_0^1 |v(t)| \, dt + \int_1^2 |v(t)| \, dt
$

$
\text{Distance} = \int_0^1 (6 - 6t) \, dt + \int_1^2 (6t - 6) \, dt
$

**First integral:**

$
\int_0^1 (6 - 6t) \, dt = \big[6t - 3t^2\big]_0^1 = (6(1) - 3(1)^2) - (6(0) - 3(0)^2) = 6 - 3 = 3
$

**Second integral:**

$
\int_1^2 (6t - 6) \, dt = \big[3t^2 - 6t\big]_1^2 = \big(3(2)^2 - 6(2)\big) - \big(3(1)^2 - 6(1)\big)
$

$
= (12 - 12) - (3 - 6) = 0 - (-3) = 3
$

**Total distance:**

$
\text{Distance} = 3 + 3 = 6
$

### Final Answers:

 **Distance traveled by the particle:** \(6\)
