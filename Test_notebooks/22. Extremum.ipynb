**6. The profit function is $ P(u) = -2u^2 + 50u - 300 $, where $ u $ is the number of units sold. Find the number of units that maximize profit.**

To maximize profit, we find the vertex of this parabola. The formula to find the vertex for a quadratic equation $ ax^2 + bx + c $ is $ u = -\frac{b}{2a} $.

Here:
- $ a = -2 $,
-  b = 50 $.

Thus:
$
u = -\frac{50}{2(-2)} = \frac{50}{4} = 12.5.
$

The profit is maximized at $ u = 12.5 $, meaning $ \approx 12.5 $ units maximize profit.

**7. You have 10 meters of string, and you need to use it to enclose the largest possible rectangular area. Find the dimensions of the rectangle.**

Let the length of the rectangle be $ l $ and the width $ w $. The perimeter of the rectangle is:
$
2l + 2w = 10 \quad \Rightarrow \quad l + w = 5 \quad \Rightarrow \quad w = 5 - l.
$


The area of the rectangle is:
$
A = l \cdot w = l(5 - l) = 5l - l^2.
$

To maximize $ A $, find the derivative and set it to zero:
$
\frac{dA}{dl} = 5 - 2l \quad \Rightarrow \quad 5 - 2l = 0 \quad \Rightarrow \quad l = 2.5.
$

If $ l = 2.5 $, then $ w = 5 - 2.5 = 2.5 $.

The rectangle with the largest area is a square with side lengths $ 2.5 \, \text{meters} $.

**8. Find the extremum of $ f(x) = x^2 + 3x - 5 $.**

Find the derivative:
$
f'(x) = 2x + 3.
$

Set $ f'(x) = 0 $ to find critical points:
$
2x + 3 = 0 \quad \Rightarrow \quad x = -\frac{3}{2}.
$

To determine if this is a minimum or maximum, use the second derivative:
$
f''(x) = 2.
$

Since $ f''(x) > 0 $, the function has a minimum at $ x = -\frac{3}{2} $.

The minimum value is:
$
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5 = \frac{9}{4} - \frac{9}{2} - 5 = -\frac{29}{4}.
$

The extremum is a **minimum** at $ x = -\frac{3}{2} $, and the minimum value is $ -\frac{29}{4} $.
