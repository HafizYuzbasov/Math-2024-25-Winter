   - $y'(x)= y(x)$

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

### **1. Solve the following first-order ordinary differential equations:**

#### (a) $ y'(x) = y(x) $

This is a separable differential equation:

$
\frac{dy}{y} = dx
$

Integrating both sides:

$
\ln|y| = x + C
$

Exponentiating both sides:

$
y = e^{x+C} = Ce^x
$

where $ C = e^C $ is an arbitrary constant.

---

#### (b) $ y'(x) = \frac{1}{2y(x)} $

Rewriting:

$
2y \, dy = dx
$

Integrating both sides:

$
y^2 = \frac{x}{2} + C
$

Taking the square root:

$
y = \pm \sqrt{\frac{x}{2} + C}
$

---

### **2. Solve the first-order ODEs using separation of variables:**

#### (a) $\frac{dy}{dx} = \frac{x}{y} $

Rewriting:

$
y \, dy = x \, dx
$

Integrating both sides:

$
\frac{y^2}{2} = \frac{x^2}{2} + C
$

Simplifying:

$
y^2 = x^2 + C_1
$

$
y = \pm \sqrt{x^2 + C_1}
$

---

#### (b) $ \frac{dy}{dx} = \frac{y}{x} $

Rewriting:

$
\frac{dy}{y} = \frac{dx}{x}
$

Integrating both sides:

$
\ln|y| = \ln|x| + C
$

$
y = Cx
$

---

#### (c) $ \frac{dy}{dx} = xy $

Rewriting:

$
\frac{dy}{y} = x \, dx
$

Integrating both sides:

$
\ln|y| = \frac{x^2}{2} + C
$

Exponentiating:

$
y = e^{\frac{x^2}{2} + C} = Ce^{\frac{x^2}{2}}
$

---

#### (d) $ \frac{dy}{dx} = xy^2 $

Rewriting:

$
\frac{dy}{y^2} = x \, dx
$

Integrating:

$
-\frac{1}{y} = \frac{x^2}{2} + C
$

Simplifying:

$
y = -\frac{1}{\frac{x^2}{2} + C}
$

### **3. Solve the second-order ordinary differential equations:**

#### (a) $ y''(x) + y'(x) = 0 $ with $ y(0) = 2 $ and $ y'(0) = -1 $

The characteristic equation is:

$
r^2 + r = 0 \quad \Rightarrow \quad r(r+1) = 0
$

Roots: $ r = 0, -1 $.

General solution:

$
y(x) = C_1 + C_2 e^{-x}
$

Using initial conditions:

1. $ y(0) = 2 $ gives $ C_1 + C_2 = 2 $.
2. $ y'(x) = -C_2 e^{-x} $, so $ y'(0) = -C_2 = -1 $, which gives $ C_2 = 1 $.

Thus, $ C_1 = 2 - 1 = 1 $, and the solution is:

$
y(x) = 1 + e^{-x}
$

---

#### (b) $ y''(x) - y(x) = 0 $ with $ y(0) = 2 $ and $ y'(0) = 0 $

The characteristic equation is:

$
r^2 - 1 = 0 \quad \Rightarrow \quad r = \pm 1
$

General solution:

$
y(x) = C_1 e^x + C_2 e^{-x}
$

Using initial conditions:

1. $ y(0) = 2 $ gives $ C_1 + C_2 = 2 $.
2. $ y'(x) = C_1 e^x - C_2 e^{-x} $, so $ y'(0) = C_1 - C_2 = 0 $, which gives $ C_1 = C_2 $.

Thus, $ C_1 = C_2 = 1 $, and the solution is:

$
y(x) = e^x + e^{-x}
$

---

#### (c) $ \frac{d^2y}{dx^2} = -\omega^2 y(x) $

The characteristic equation is:

$
r^2 + \omega^2 = 0 \quad \Rightarrow \quad r = \pm i\omega
$

General solution:

$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$

### **4. Verify $ \psi(t, x) = A \cos(\omega t + kx) $ is a solution of the wave equation**

The wave equation is:

$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0
$

Compute $ \frac{\partial^2 \psi}{\partial t^2} $:

$
\psi = A \cos $ omega t + kx $
$
$
\frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx)
$
$
\frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
$

Compute $ \frac{\partial^2 \psi}{\partial x^2} $:

$
\frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx)
$
$
\frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx)
$

Substitute into the wave equation:

$
-A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0
$

Simplify:

$
-A (\omega^2 - v^2 k^2) \cos(\omega t + kx) = 0
$

Since $ \cos(\omega t + kx) \neq 0 $, we require:

$
\omega^2 = v^2 k^2 \quad \Rightarrow \quad v = \frac{\omega}{k}
$

Thus, $ \psi(t, x) $ satisfies the wave equation.
