# Lecture Notes: Lines in the Plane

## 1. Types of Line Equations in the Plane

In the plane, a line can be represented using different types of equations, each with its specific characteristics. The most common forms of line equations are:

- **General equation of a line**
- **Canonical equation of a line**
- **Intercept form of the equation**
- **Slope-intercept form of the equation**
- **Direction form of the equation**
- **Equation of a line passing through two given points**

Let's discuss each of these in detail.

## 2. General Equation of a Line

The general equation of a line in the Cartesian plane is given by:

$$
Ax + By + C = 0
$$

Where:
- $A$, $B$, and $C$ are constants.
- $A$ and $B$ are not both zero.

This form is called the **general form** of the line. It is a linear equation in two variables, $x$ and $y$.

### Example:

$$
2x - 3y + 5 = 0
$$

This represents a line with a slope and intercept that can be derived after rearranging it into other forms (e.g., slope-intercept form).

## 3. Canonical Equation of a Line

The canonical form (or normal form) of a line represents the line in terms of its **distance from the origin** and the **angle it makes with the positive x-axis**. The equation is given by:

$$
x \cos \theta + y \sin \theta = p
$$

Where:
- $\theta$ is the angle between the line and the positive x-axis.
- $p$ is the perpendicular distance from the origin to the line.

### Example:

$$
x \cos 30^\circ + y \sin 30^\circ = 5
$$

This equation describes a line in terms of its normal distance from the origin.

## 4. Line Equation in Intercept Form on the Coordinate Axes

In the intercept form, the line is expressed in terms of its intercepts on the x-axis and y-axis. The equation is given by:

$$
\frac{x}{a} + \frac{y}{b} = 1
$$

Where:
- $a$ is the x-intercept (the point where the line intersects the x-axis).
- $b$ is the y-intercept (the point where the line intersects the y-axis).

### Example:

$$
\frac{x}{3} + \frac{y}{4} = 1
$$

This represents a line that intersects the x-axis at $(3, 0)$ and the y-axis at $(0, 4)$.

## 5. Line Equation in Slope-Intercept Form

The slope-intercept form of the line is one of the most commonly used equations, particularly in geometry and algebra. It is given by:

$$
y = mx + c
$$

Where:
- $m$ is the **slope** of the line.
- $c$ is the **y-intercept** (the point where the line crosses the y-axis).

### Example:

$$
y = 2x + 3
$$

This represents a line with a slope of $2$ and a y-intercept of $3$.

## 6. Line Equation Passing Through a Given Point in a Given Direction

If a line passes through a given point $(x_1, y_1)$ and has a direction vector $(a, b)$, the equation of the line can be written as:

$$
\frac{x - x_1}{a} = \frac{y - y_1}{b}
$$

This form is called the **direction form** of the line.

### Example:

If the line passes through $(1, 2)$ and has direction vector $(3, 4)$, the equation is:

$$
\frac{x - 1}{3} = \frac{y - 2}{4}
$$

## 7. Line Equation Passing Through Two Given Points

If two points $(x_1, y_1)$ and $(x_2, y_2)$ are given, the equation of the line passing through them can be derived using the **point-slope form**. The slope $m$ of the line is given by:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Then, the equation of the line is:

$$
y - y_1 = m(x - x_1)
$$

### Example:

For points $(1, 2)$ and $(3, 4)$, the slope is:

$$
m = \frac{4 - 2}{3 - 1} = 1
$$

Thus, the equation of the line is:

$$
y - 2 = 1(x - 1) \quad \text{or} \quad y = x + 1
$$

## 8. Relative Positions of Lines in the Plane

The relative position of two lines in the plane can be classified into the following categories:

1. **Intersecting lines**: Two lines that cross at a point.
2. **Parallel lines**: Two lines that never meet, having the same slope.
3. **Coincident lines**: Two lines that lie exactly on top of each other (identical lines).

## 9. Intersection of Two Lines

To find the intersection point of two lines, we solve the system of equations representing the lines. Suppose we have the following two equations:

$$
A_1x + B_1y + C_1 = 0
$$
$$
A_2x + B_2y + C_2 = 0
$$

We solve these two equations simultaneously to get the values of $x$ and $y$ that satisfy both equations. This solution will give us the coordinates of the intersection point.

### Example:

Solve the system of equations:

$$
2x + 3y = 6
$$
$$
4x - y = 5
$$

## 10. Angle Between Two Lines

The angle $\theta$ between two lines with slopes $m_1$ and $m_2$ is given by the formula:

$$
\tan \theta = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|
$$

Where $\theta$ is the angle between the two lines. This formula gives the acute angle between the lines.

### Example:

For lines with slopes $m_1 = 2$ and $m_2 = -1$, the angle is:

$$
\tan \theta = \left| \frac{2 - (-1)}{1 + 2(-1)} \right| = \left| \frac{3}{-1} \right| = 3
$$

Then, $\theta = \tan^{-1}(3)$.

## 11. Conditions for Parallelism and Perpendicularity of Two Lines

- **Parallel Lines**: Two lines are parallel if their slopes are equal. That is, if $m_1 = m_2$.
  
  Example: The lines $y = 2x + 1$ and $y = 2x - 3$ are parallel because both have slope $m = 2$.

- **Perpendicular Lines**: Two lines are perpendicular if the product of their slopes is $-1$. That is, if $m_1 \cdot m_2 = -1$.
  
  Example: The lines $y = 2x + 1$ and $y = -\frac{1}{2}x - 3$ are perpendicular because $2 \cdot (-\frac{1}{2}) = -1$.
