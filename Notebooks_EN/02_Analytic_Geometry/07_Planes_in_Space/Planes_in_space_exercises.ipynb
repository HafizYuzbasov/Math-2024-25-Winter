$$

$$





$$
\vec{n}_1 = (1, 1, 1)
$$



$$
\vec{n}_2 = (1, 0, 0)
$$



$$
\cos \theta = \frac{\vec{n}_1 \cdot \vec{n}_2}{|\vec{n}_1| |\vec{n}_2|}
$$




$$
\vec{n}_1 \cdot \vec{n}_2 = (1)(1) + (1)(0) + (1)(0) = 1
$$



$$
|\vec{n}_1| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}
$$



$$
|\vec{n}_2| = \sqrt{1^2 + 0^2 + 0^2} = 1
$$



$$
\cos \theta = \frac{1}{\sqrt{3} \times 1} = \frac{1}{\sqrt{3}}
$$



$$
\theta = \cos^{-1}\left( \frac{1}{\sqrt{3}} \right)
$$


$$
\cos^{-1}\left( \frac{1}{\sqrt{3}} \right) \approx 54.74^\circ
$$



$$
\theta \approx 54.74^\circ
$$
