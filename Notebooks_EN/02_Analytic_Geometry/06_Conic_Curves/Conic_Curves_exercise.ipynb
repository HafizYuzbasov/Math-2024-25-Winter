```python
# Import necessary libraries
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define variables
x, y = sp.symbols('x y')

# Circle's center and radius
center = (1, 2)
radius = 3

# Circle equation
circle_eq = (x - center[0])**2 + (y - center[1])**2 - radius**2
print("Equation of the circle:", sp.simplify(circle_eq), "= 0")

# Plot the circle
theta = np.linspace(0, 2 * np.pi, 500)
x_circle = center[0] + radius * np.cos(theta)
y_circle = center[1] + radius * np.sin(theta)

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x_circle, y_circle, label=f"Circle: Center={center}, Radius={radius}")
plt.scatter(center[0], center[1], color='red', label="Center (1, 2)")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.axis('equal')  # Equal scaling for x and y
plt.title("Graph of the Circle")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def plot_parabola():
    x_vals = np.linspace(1, 5, 500)
    a = -1  # From calculations
    parabola_y = a * (x_vals - 2) * (x_vals - 4)

    plt.figure(figsize=(6, 6))
    plt.plot(x_vals, parabola_y, label="Parabola: Intersects x=2, x=4")
    plt.scatter(3, 1, color='red', label="Point (3, 1)")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.title("Graph of the Parabola")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

plot_parabola()
```

```python
def plot_ellipse():
    x_vals = np.linspace(-3, 5, 500)
    y_vals = np.linspace(-3, 5, 500)
    x_grid, y_grid = np.meshgrid(x_vals, y_vals)

    # Equation of ellipse
    ellipse_eq = x_grid**2 + 4*y_grid**2 - 4*x_grid - 16*y_grid + 16

    plt.figure(figsize=(6, 6))
    plt.contour(x_grid, y_grid, ellipse_eq, levels=[0], colors='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title("Graph of the Ellipse")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis('equal')
    plt.show()

plot_ellipse()
```

```python
def plot_tangent_and_circle():
    theta = np.linspace(0, 2 * np.pi, 500)
    x_circle = np.cos(theta)
    y_circle = np.sin(theta)

    # Tangent line y = mx - 5 (calculated slope m = sqrt(6)/3)
    m = sp.sqrt(6) / 3
    x_vals = np.linspace(-1.5, 1.5, 500)
    tangent_y = m * x_vals - 5

    plt.figure(figsize=(6, 6))
    plt.plot(x_circle, y_circle, label="Circle: x^2 + y^2 = 1")
    plt.plot(x_vals, tangent_y, label="Tangent Line", color='red')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.title("Tangent to Circle")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis('equal')
    plt.show()

plot_tangent_and_circle()
```

```python
def plot_hyperbola_and_line():
    x_vals = np.linspace(-3, 3, 500)
    y_vals = np.linspace(-3, 3, 500)
    x_grid, y_grid = np.meshgrid(x_vals, y_vals)

    # Hyperbola x^2 - y^2 = 1
    hyperbola_eq = x_grid**2 - y_grid**2 - 1

    # Line x^2 + 4y^2 = 6
    ellipse_line_eq = x_grid**2 + 4*y_grid**2 - 6

    plt.figure(figsize=(6, 6))
    plt.contour(x_grid, y_grid, hyperbola_eq, levels=[0], colors='blue', label="Hyperbola")
    plt.contour(x_grid, y_grid, ellipse_line_eq, levels=[0], colors='green', label="Line")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title("Hyperbola and Intersection Line")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis('equal')
    plt.show()

plot_hyperbola_and_line()
```

```python
def plot_hyperbola_distance():
    x_vals = np.linspace(-3, 3, 500)
    y_vals = np.linspace(-3, 3, 500)
    x_grid, y_grid = np.meshgrid(x_vals, y_vals)

    # Hyperbola x^2 - y^2 = 1
    hyperbola_eq = x_grid**2 - y_grid**2 - 1

    plt.figure(figsize=(6, 6))
    plt.contour(x_grid, y_grid, hyperbola_eq, levels=[0], colors='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title("Distance Between Hyperbola Branches")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis('equal')

    # Distance between branches = 2 * a (here, a = 1)
    plt.text(0, 0, "Distance = 2", fontsize=12, color="red")
    plt.show()

plot_hyperbola_distance()
```
